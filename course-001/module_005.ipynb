### Creating a graphic

```python
from random import randrange
import matplotlib.pyplot as plt
```

```python
grades = []

for i in range(10):

  grades.append(randrange(0, 11))

grades
```

```python
x_axis = list(range(1, 11))
x_axis
```

```python
y_axis = grades
y_axis
```

```python
plt.plot(x_axis, y_axis, marker = 'o')
plt.title('Math grades')
plt.xlabel('Exames')
plt.ylabel('Grades')
plt.grid()
plt.show()
```

```python
math_grades = ['Math', 8, 7, 6, 10, 7, 7, 8, 10]
physics_grades = ['Physics', 10, 10, 6, 7, 7, 7, 8, 7]

grades = [math_grades, physics_grades]

for grade in grades:

  x_axis = list(range(1, 9))
  y_axis = grade[1:]
  
  plt.plot(x_axis, y_axis, marker = 'o')
  plt.xlabel('Exames')
  plt.ylabel('Grades')
  plt.title(grade[0])
  plt.grid()
  plt.show()

  print('')
```
